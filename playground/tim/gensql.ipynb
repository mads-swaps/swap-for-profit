```python
cols = ['open', 'high', 'low', 'close']
shifts = list(range(1,15))

sql_strs = []

for l in shifts:
    for c in cols:
        sql_strs.append(f"""	LAG(ch.{c},{l}) OVER (ORDER BY ch.open_time) {c}_{l},
""")
        
        
print("".join(sql_strs))

sql_strs = []

for l in shifts:
    for c in cols:
        sql_strs.append(f"""new.{c}_{l}, """)
        
        
print("".join(sql_strs))
```
