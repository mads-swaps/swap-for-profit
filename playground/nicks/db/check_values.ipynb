# Check Column Differences

```python
# Column comparison
ignore_columns = ['pair_id','open_time','open','high','low','close','volume','quote_asset_volume',
                  'number_of_trades','taker_buy_base_asset_volume','taker_buy_quote_asset_volume',
                  'close_time','created_at','updated_at'
                 ]

for col in feature_df.columns:
    if col not in ignore_columns:
        if col not in compare_df.columns:
            print(f'Additional column in features: {col}')
for col in compare_df.columns:
    if col not in ignore_columns:
        if col not in feature_df.columns:
            print(f'Missing column in features: {col}')
```

# Check Value Differences

```python
# Value comparison
import numpy as np

merged_df = feature_df.merge(compare_df, how='left', left_on='open_time', right_on='open_time')
check_cols = [col[:-2]for col in merged_df.columns if col.endswith('_x')]

for col in check_cols:
    if any(~np.isclose(merged_df[f'{col}_x'], merged_df[f'{col}_y'])):
        percent = (merged_df[f'{col}_x'] != merged_df[f'{col}_y']).sum() / len(merged_df)
        difference = np.mean(abs(merged_df[f'{col}_x'] - merged_df[f'{col}_y']))
        example = merged_df[merged_df[f'{col}_x'] != merged_df[f'{col}_y']].sample(1)[['open_time', f'{col}_x',f'{col}_y']].values
        print(f'Column {col} is off on {percent} of values by a mean difference of {difference}')
        print(f'  Example = {example}\n')
```

```python
def calculate_atr(df, period=14):
    '''
    Calculate the Average True Range (ATR)
    '''
    df2=df[['high','low','close','close_1']].copy()
    df2['tr'] = np.max([df2.high, df2.close_1], axis=0) - np.min([df2.low, df2.close_1], axis=0)
    df2['atr'] = df2.tr.rolling(period).mean()
    return df2[['high','close_1','low','tr','atr']]

calculate_atr(feature_df.loc[59-13:59].set_index('open_time')) #.iloc[-1:]
```
